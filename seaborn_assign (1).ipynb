Seaborn is a powerful data visualization library in Python that offers various plot types. Here are five commonly used plots and their uses:

1. **Scatter Plot (`sns.scatterplot()`):**
   - Use: Scatter plots are used to visualize the relationship between two continuous variables. Each data point is represented as a point on the plot, and it helps to identify patterns, clusters, or correlations between the variables.

2. **Bar Plot (`sns.barplot()`):**
   - Use: Bar plots are used to display and compare the values of different categories. They are suitable for visualizing categorical data, showing the mean or aggregate values of each category.

3. **Histogram Plot (`sns.histplot()`):**
   - Use: Histograms are used to visualize the distribution of a single numerical variable. It shows the frequency or count of data points falling within specific bins, allowing you to understand the data's underlying distribution.

4. **Box Plot (`sns.boxplot()`):**
   - Use: Box plots, also known as box-and-whisker plots, are used to display the distribution of data and identify potential outliers. They show the median, quartiles, and possible extreme values in the data.

5. **Heatmap (`sns.heatmap()`):**
   - Use: Heatmaps are used to visualize the correlation between variables in a tabular form. It uses colors to represent the strength of the relationship between variables, making it easy to spot patterns and dependencies.

QUESTION 3

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Load the "titanic" dataset
titanic_data = sns.load_dataset("titanic")

#  Create the first box plot for 'pclass' vs. 'age'
plt.figure(figsize=(8, 6))
sns.boxplot(x="pclass", y="age", data=titanic_data)
plt.title("Box Plot of 'pclass' vs. 'age'")
plt.show()

#  Create the second box plot for 'pclass' vs. 'fare'
plt.figure(figsize=(8, 6))
sns.boxplot(x="pclass", y="fare", data=titanic_data)
plt.title("Box Plot of 'pclass' vs. 'fare'")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights_data = sns.load_dataset("flights")

# Pivot the dataset to prepare for the heatmap
flights_pivot = flights_data.pivot("month", "year", "passengers")

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="coolwarm", linewidths=0.5)
plt.title("Flights Dataset Heatmap")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```
